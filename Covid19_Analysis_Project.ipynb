```python
# Importing Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading COVID-19 dataset
covid_data = pd.read_csv("covid19_Confirmed_dataset.csv")
covid_data.head()
```

```python
# Checking the shape of dataset
covid_data.shape
```

```python
# Deleting unrequired columns
df = covid_data.drop(["Lat", "Long"], axis = 1, inplace = True)
```

```python
# Rechecking the columns
covid_data.head()
```

```python
# Aggregate the rows by country
covid_data_aggregated = covid_data.groupby("Country/Region").sum()
```

```python
covid_data_aggregated.head()
```

```python
# Checking for Shape
covid_data_aggregated.shape
```

```python
# Visualise data related to a country
covid_data_aggregated.loc["Spain"].plot()
covid_data_aggregated.loc["India"].plot()
covid_data_aggregated.loc["Pakistan"].plot()
plt.legend()
```

```python
# Calculating a good measure
covid_data_aggregated.loc["India"].plot()
```

```python
# Calculating a good measure for first 3 days
covid_data_aggregated.loc["India"][:30].plot()
```

```python
# Calculate the first derivative of the curve
covid_data_aggregated.loc["India"].diff().plot()
```

```python
# Calculating the maximum infection rate
covid_data_aggregated.loc["India"].diff().max()
```

```python
covid_data_aggregated = covid_data_aggregated.apply(pd.to_numeric, errors='coerce')
```

```python
countries = list(covid_data_aggregated.index)
max_infection_rates = []

for c in countries:
    max_infection_rates.append(covid_data_aggregated.loc[c].diff().max())
covid_data_aggregated["Maximum_Infection_Rates"] = max_infection_rates
```

```python
covid_data_aggregated
```

```python
# Creating a new dataframe
covid_data_new = pd.DataFrame(covid_data_aggregated["Maximum_Infection_Rates"])
```

```python
covid_data_new
```

```python
# Importing the World_Happiness dataset
happiness_report = pd.read_csv("worldwide_happiness_report.csv")
```

```python
happiness_report
```

```python
# Dropping the unrequired columns
unrequired_columns = ["Overall rank", "Score", "Generosity", "Perceptions of corruption"]
```

```python
happiness_report.drop(unrequired_columns, axis = 1, inplace = True)
happiness_report.head()
```

```python
# Setting Index as Country or region
happiness_report.set_index("Country or region", inplace = True)
happiness_report.head()
```

```python
# checking shape of the first dataset
covid_data_new.shape
```

```python
# checking shape of the second dataset
happiness_report.shape
```

```python
# Joining both the dataset
data = covid_data_new.join(happiness_report, how = "inner")
data
```

```python
# checking corelation matrix
data.corr()
```

```python
data
```

```python
# Visualizing Scatterplot between GDP and Maximum infection rate
x = data["GDP per capita"]
y = data["Maximum_Infection_Rates"]
sns.scatterplot(x=x, y= np.log(y))
plt.xlabel("GDP per capita")
plt.ylabel("Maximum Infection Rates")
plt.title("Scatterplot: GDP vs. Infection Rates")
plt.show()
```

```python
# Regression Plot
sns.regplot(x=x, y= np.log(y))
plt.xlabel("GDP per capita")
plt.ylabel("Maximum Infection Rates")
plt.title("Regression Plot: GDP vs. Infection Rates")
plt.show()
```
