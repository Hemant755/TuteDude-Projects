```python
# Rolling two dice to calculate the probability of getting a sum of 7
def dice_sum_simulation():
    count = 0
    for _ in range(10000):  # 10,000 trials
        die1 = random.randint(1, 6)
        die2 = random.randint(1, 6)
        if die1 + die2 == 7:
            count += 1
    print("Probability of getting a sum of 7:", count / 10000)

coin_toss_simulation()
```

```python
def at_least_one_six():
    success = 0
    for _ in range(10000):  # 10,000 trials
        rolls = [random.randint(1, 6) for _ in range(10)]
        if 6 in rolls:
            success += 1
    print("Probability of at least one '6':", success / 10000)

at_least_one_six()
```

```python
def bayes_theorem_simulation():
    red, green, blue = 5, 7, 8
    total_balls = red + green + blue
    count_red_given_blue = 0
    blue_prev = 0

    for _ in range(1000):  # 1,000 trials
        first_draw = random.choices(["Red", "Green", "Blue"], [red, green, blue])[0]
        second_draw = random.choices(["Red", "Green", "Blue"], [red, green, blue])[0]
        if first_draw == "Blue":
            blue_prev += 1
            if second_draw == "Red":
                count_red_given_blue += 1

    prob_red_given_blue = count_red_given_blue / blue_prev if blue_prev > 0 else 0
    print("P(Red | Blue):", prob_red_given_blue)

bayes_theorem_simulation()
```

```python
import numpy as np

def discrete_random_variable():
    sample = np.random.choice([1, 2, 3], size=1000, p=[0.25, 0.35, 0.4])
    mean = np.mean(sample)
    variance = np.var(sample)
    std_dev = np.std(sample)

    print("Empirical Mean:", mean)
    print("Empirical Variance:", variance)
    print("Empirical Standard Deviation:", std_dev)

discrete_random_variable()
```

```python
import matplotlib.pyplot as plt

def exponential_distribution():
    samples = np.random.exponential(scale=5, size=2000)  # Mean = 5
    plt.hist(samples, bins=30, density=True, alpha=0.6, color='g')

    # Overlay PDF
    x = np.linspace(0, max(samples), 1000)
    pdf = (1/5) * np.exp(-x / 5)  # PDF for exponential distribution with mean=5
    plt.plot(x, pdf, 'r-', lw=2)

    plt.title('Exponential Distribution')
    plt.xlabel('Value')
    plt.ylabel('Density')
    plt.show()

exponential_distribution()
```

```python
def central_limit_theorem():
    uniform_samples = np.random.uniform(0, 1, 10000)  # 10,000 random numbers
    plt.hist(uniform_samples, bins=30, alpha=0.6, color='Blue', label='Uniform Distribution')

    # Sample means
    sample_means = [np.mean(np.random.choice(uniform_samples, 30)) for _ in range(1000)]
    plt.hist(sample_means, bins=30, alpha=0.6, color='Red', label='Sample Mean Distribution')

    plt.title('Central Limit Theorem Simulation')
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.legend()
    plt.show()

central_limit_theorem()
```

```python
# Simulates the Central Limit Theorem by drawing 10,000 samples from a uniform distribution.
# Computes sample means from 1,000 subsets of size 30.
# Compares the uniform distribution to the sample mean distribution to illustrate the theorem.
```
