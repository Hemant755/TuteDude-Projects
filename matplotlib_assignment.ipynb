```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [10, 15, 25, 30, 50]

# a. Plot the data
plt.plot(x, y, marker='o')

# b. Customize the plot
plt.title("Simple Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)

# Show the plot
plt.show()
```

```python
# Data
students = ['John', 'Jane', 'Alice', 'Bob']
marks = [75, 85, 60, 90]

# a. Plot the data
plt.bar(students, marks, color=['blue', 'green', 'orange', 'red'])

# b. Customize the plot
plt.title("Marks Scored by Students")
plt.xlabel("Students")
plt.ylabel("Marks")

# Show the plot
plt.show()
```

```python
# Data
regions = ['North America', 'Europe', 'Asia', 'Others']
revenue = [45, 25, 20, 10]

# a. Create the pie chart
explode = [0.1 if r == max(revenue) else 0 for r in revenue]  # Highlight the highest revenue region
plt.pie(revenue, labels=regions, autopct='%1.1f%%', explode=explode, startangle=140)

# b. Add a title
plt.title("Revenue Distribution by Region")

# Show the plot
plt.show()
```

```python
import numpy as np

# Data: Randomly generate 1000 integers between 1 and 100
data = np.random.randint(1, 101, size=1000)

# Plot the histogram
plt.hist(data, bins=10, color='purple', edgecolor='black')

# Customize the plot
plt.title("Frequency Distribution")
plt.xlabel("Range")
plt.ylabel("Frequency")

# Show the plot
plt.show()
```
